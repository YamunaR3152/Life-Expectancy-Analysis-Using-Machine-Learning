```python
#STEP : 01 IMPORT LIBRABIES AND LOAD DATASET 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import 
# Load the cleaned dataset
df = pd.read_csv(r"C:/Users/yamun/OneDrive/Desktop/Life Expectancy Data.csv")

# Display first few rowsseaborn as sns

df.head()
```

```python
#Step 2: Basic dataset information
# Shape of dataset
print("Shape of the dataset:", df.shape)

print("\nData types:")


# Data types and non-null values
print("\nDataset Info:")
print(df.info())

```

```python

# Check missing values before filling
print("ðŸ” Missing values before cleaning:\n", df.isnull().sum())

# Separate numeric and categorical columns
num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns

# Fill missing values
# â†’ Numeric columns â†’ median
df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.median()))

# â†’ Categorical columns â†’ mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Check again
print("\nâœ… Missing values after cleaning:\n", df.isnull().sum())

# Handle duplicates
duplicates = df.duplicated().sum()
print(f"\nðŸ§¹ Duplicate rows found: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    print("âœ… Duplicates removed.")
else:
    print("No duplicate rows found.")

# Convert GDP to numeric (if stored as string)
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')

# Final check
print("\nâœ… Final missing values count:", df.isnull().sum().sum())
```

```python


# Find outliers using IQR method for numeric columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    print(f"{col}: {len(outliers)} outliers")
```
